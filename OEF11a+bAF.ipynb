```python
# x,y,z,r uitrekennen 
EQ1 = Eq((-3-x)**2+(3-y)**2+(-4-z)**2 - r**2 )
EQ2 = Eq((1-x)**2+(1-y)**2+(z)**2-r**2)
EQ3 = Eq((3-x)**2+(1-y)**2+(-2-z)**2 -r**2)
EQ4 = Eq((-5-x)**2+(5-y)**2+(-4-z)**2 -r**2)
solve([EQ1,EQ2,EQ3,EQ4], (x, y, z, r))
# hiermee hebben we de vergelijking van de bol : (x-2)**2 + (y-10)**2 +(z+1)**2 = 83
# bij controle door de punten in te geven komen we 4 geldige vergelijkingen uit, en in geogebra klopt het ook.
# de straal is gelijk aan sqrt(83)
# het middelpunt is gelijk aan (2,10,-1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Middelpunt en straal van de bol
center = (2, 10, -1)
radius = np.sqrt(83)

# Maak een grid van punten voor de bol
u = np.linspace(0, 2 * np.pi, 100)  # Hoek in de xy-vlak
v = np.linspace(0, np.pi, 100)       # Hoek in de yz-vlak

# Parametriseer de bol
x = center[0] + radius * np.outer(np.cos(u), np.sin(v))
y = center[1] + radius * np.outer(np.sin(u), np.sin(v))
z = center[2] + radius * np.outer(np.ones(np.size(u)), np.cos(v))

# Coördinaten van de punten A, B, C en D
points = {
    'A': (-3, 3, 4),
    'B': (3, 1, -2),
    'C': (1, 1, 0),
    'D': (-5, 5, 0)
}

# Maak een 3D-plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot de bol
ax.plot_surface(x, y, z, color='w', alpha=0.6)  # Kleur en transparantie

# Plot de punten A, B, C en D
for point_name, (px, py, pz) in points.items():
    ax.scatter(px, py, pz, label=point_name, s=100)  # s = grootte van de marker

# Labels en titel
ax.set_xlabel('X-as')
ax.set_ylabel('Y-as')
ax.set_zlabel('Z-as')
ax.set_title('3D Bol met middelpunt (2, 10, -1) en straal √83')

# Toon de legenda
ax.legend()

# Toon de plot
plt.show()
```

```python
#Oefening 11a

# 1->afstandsformule van vlak tot middelpunt (straal): r = |Ax+By+Cz+D|/sqrt(A**2+ B**2+ C22)
                                                    #  r=  |3*1+1+2*2+6|/sqrt(3**2+(-1)**2+2**2)
                                                    #  r=  sqrt(14)
        
A = 3
B = -1
C = 2
D = 6
    
x=1
y=-1
z=2
    
R= (abs(A*x+B*y+C*z+D)/sqrt(A**2+B**2+C**2))

# 2-> formule bol opstellen met M en r: (x-1)**2 + (y+1)**2 + (z-2)**2 = 14

# 3 -> Grafisch voorstellen:


import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sympy import symbols, Eq, solve

# Definieer de variabelen
x, y, z = symbols('x y z')

# De vergelijking van het vlak: bijvoorbeeld 3x - y + 2z + 6 = 0
plane_eq = Eq(3*x - y + 2*z + 6, 0)

# Oplossen voor z zodat we het vlak kunnen plotten
z_plane = solve(plane_eq, z)[0]

# Functie om het vlak te plotten
def plot_plane(ax):
    x_vals = np.linspace(-10, 10, 100)
    y_vals = np.linspace(-10, 10, 100)
    X, Y = np.meshgrid(x_vals, y_vals)
    Z = (6 - 3*X + Y) / 2  # z = (6 - 3x + y) / 2
    ax.plot_surface(X, Y, Z, color='yellow', alpha=0.5)

# Middelpunt en straal van de bol
M = np.array([1, -1, 2])  # Geef het middelpunt op, bijvoorbeeld (1, -1, 2)
r = np.sqrt(14)  # Straal sqrt(14) of de juiste waarde op basis van jouw berekening

# Functie om de bol te plotten
def plot_sphere(ax):
    u = np.linspace(0, 2 * np.pi, 100)
    v = np.linspace(0, np.pi, 100)

    # Parametriseer de bol in 3D
    x_sphere = M[0] + r * np.outer(np.cos(u), np.sin(v))
    y_sphere = M[1] + r * np.outer(np.sin(u), np.sin(v))
    z_sphere = M[2] + r * np.outer(np.ones(np.size(u)), np.cos(v))

    # Plot de bol
    ax.plot_surface(x_sphere, y_sphere, z_sphere, color='blue', alpha=0.6, rstride=5, cstride=5)

# Maak een 3D-plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot het vlak
plot_plane(ax)

# Plot de bol
plot_sphere(ax)

# Labels en titel
ax.set_xlabel('X-as')
ax.set_ylabel('Y-as')
ax.set_zlabel('Z-as')
ax.set_title('Bol en vlak')

# Toon de plot
plt.show(), R
```
